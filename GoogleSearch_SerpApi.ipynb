```python
import os
api_key = os.environ.get('SerpApi')
```

```python
from serpapi import GoogleSearch
search = GoogleSearch({
    "q": "coffee", 
    "location": "Austin,Texas",
    "api_key": api_key
  })
result = search.get_dict()
```

```python
from serpapi import GoogleSearch

params = {
  "engine": "google_news",
  "q": "pizza",
  "api_key": api_key
}

search = GoogleSearch(params)
results = search.get_dict()
news_results = results["news_results"]
news_results
```
